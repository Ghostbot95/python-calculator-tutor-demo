```python
""""
Python Calculator (Beginner Friendly Project)

In this notebook, we build a simple command-line calculator using Python.

Concepts covered:
- Functions
- User input
- Error handling
- Loops and conditional statements*/
"""

#Arithmetic Functions
def add(x, y):
    return x + y
def subtract(x, y):
    return x - y
def multiply(x, y):
    return x * y
def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y
#Main Function & Menu Setup
def main():
    print("Python Calculator")
    print("Select operation:")
    print("1. Add\n2. Subtract\n3. Multiply\n4. Divide")
    #Infinite Loop & User Choice
    while True:
        choice = input("\nEnter choice (1/2/3/4) or 'q' to quit: ")
        if choice.lower() == 'q':
            break
        #Input Validation with try-except
        if choice in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                print("Invalid input. Please enter numeric values.")
                continue
            #Conditional Logic & Function Calls
            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")
            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")
            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")
            elif choice == '4':
                print(f"{num1} / {num2} = {divide(num1, num2)}")
        #Invalid Choice Handling
        else:
            print("Invalid Input")
#Program Entry Point
main()
```
